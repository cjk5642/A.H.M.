```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

arenado = pd.read_csv(r'C:\Users\colli\FOLDER OF ALL\Desktop\Projects\DataForBaseBall\Arenado, Harper, Machado\ArenadoData.csv')
arenado = arenado.drop(columns=['OPS+','GDP','SH','SF','IBB'])
arenado['Name'] = 'Arenado'

harper = pd.read_csv(r'C:\Users\colli\FOLDER OF ALL\Desktop\Projects\DataForBaseBall\Arenado, Harper, Machado\HarperData.csv')
harper = harper.drop(columns=['OPS+','GDP','SH','SF','IBB'])
harper['Name'] = 'Harper'

machado = pd.read_csv(r'C:\Users\colli\FOLDER OF ALL\Desktop\Projects\DataForBaseBall\Arenado, Harper, Machado\MachadoData.csv')
machado = machado.drop(columns=['OPS+','GDP','SH','SF','IBB'])
machado['Name'] = 'Machado'
```

```python
total = pd.concat([arenado, harper, machado], ignore_index=True)
total = total.groupby(['Year','Name']).mean()
total = total.drop(columns=['PA','BB','AB','2B','3B','HR','RBI','SB','CS','SO','TB','HBP','H','R']) #.sort_values(by='Year')
display(np.round(total, decimals=3))
```

```python
years = [2013, 2014, 2015, 2016, 2017, 2018]

aBA = arenado.loc[:, 'BA']
aOBP = arenado.loc[:, 'OBP']
aSLG = arenado.loc[:, 'SLG']
aOPS = arenado.loc[:, 'OPS']

hBA = harper.loc[:, 'BA']
hOBP = harper.loc[:, 'OBP']
hSLG = harper.loc[:, 'SLG']
hOPS = harper.loc[:, 'OPS']

mBA = machado.loc[:, 'BA']
mOBP = machado.loc[:, 'OBP']
mSLG = machado.loc[:, 'SLG']
mOPS = machado.loc[:, 'OPS']

labels=['Arenado', 'Harper','Machado']
bottom_labels = ['BA', 'OBP', 'SLG', 'OPS']

for i in range(0, len(years)):
    plt.title("Stats for " + str(years[i]), fontsize=24)
    
    y1 = [aBA[i], aOBP[i], aSLG[i], aOPS[i] ]
    y2 = [hBA[i], hOBP[i], hSLG[i], hOPS[i] ]
    y3 = [mBA[i], mOBP[i], mSLG[i], mOPS[i] ]
    
    x = np.arange(len(bottom_labels))
    bar_width=0.25
    
    plt.bar(x, y1, bar_width, color='purple', edgecolor='black')
    plt.bar(x+bar_width, y2, bar_width, color='red', edgecolor='black')
    plt.bar(x+bar_width+bar_width, y3, bar_width, color='blue', edgecolor='black')

    
    plt.xticks(x, labels=bottom_labels, fontsize=16)
    plt.legend(labels)
    plt.show()
```
